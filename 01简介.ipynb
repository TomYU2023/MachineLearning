### 相关词汇：
  
特征：自变量  
标签：因变量  

训练集：用于训练机器学习模型的数据集。一般占70%-80%的数据集。  
验证集：用于调整模型超参数和评估模型性能的数据集，可以帮助选择最佳模型。  
测试集：用于评估最终模型性能的数据集，模型在测试集上的表现可以预测其在真实环境中的表现。一般占20%-30%的数据集。  
过拟合：模型在训练集上表现很好，但在验证集或测试集上表现较差的现象，表示模型过度学习了训练集的细节，而忽略了泛化能力。  
欠拟合：模型无法在训练集上达到很好的表现，也无法在验证集或测试集上实现良好的性能，表示模型过于简单或者没有充分学习数据的特征。  
过拟合和欠拟合是常见的机器学习问题，可以通过调整模型复杂度、增加训练数据、使用正则化等方法来解决。  

### 模型评估指标：

准确率：正确预测的样本数占总样本数的比例  
召回率：正确预测的正例样本数占总正例样本数的比例  
F1分数：准确率和召回率的调和平均值  
AUC：ROC曲线下面积，表示模型在分类任务中的性能，AUC越高，模型性能越好  

泛化能力：模型在测试集上的表现  
奥卡姆剃刀：模型越简单，泛化能力越强  
测试误差：模型在测试集上的表现
训练误差：模型在训练集上的表现


```python
import numpy as np
import matplotlib.pyplot as plt

# 生成模拟数据
np.random.seed(0)
X = np.linspace(0, 5, 20)
y = 1 + 2*X + np.random.normal(0, 1, 20)

# 拟合不同阶数的多项式模型
degrees = [1, 10]

plt.figure(figsize=(10, 5))

# 绘制欠拟合图
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='black')

degree = degrees[0]
coeffs = np.polyfit(X, y, degree)
poly = np.poly1d(coeffs)
plt.plot(X, poly(X), label=f'Degree {degree}', color='blue')

plt.title('Underfitting')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()

# 绘制过拟合图
plt.subplot(1, 2, 2)
plt.scatter(X, y, color='black')

degree = degrees[1]
coeffs = np.polyfit(X, y, degree)
poly = np.poly1d(coeffs)
plt.plot(X, poly(X), label=f'Degree {degree}', color='red')

plt.title('Overfitting')
plt.xlabel('X')
plt.ylabel('y')
plt.ylim([-5, 20])
plt.legend()

plt.tight_layout()
plt.show()
```
